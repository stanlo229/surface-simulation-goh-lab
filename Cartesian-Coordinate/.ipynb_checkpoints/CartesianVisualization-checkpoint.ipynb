```python
%matplotlib qt
import simpy
import statistics
import random
import matplotlib
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy.linalg as la
import math
import copy
from tabulate import tabulate
'''
Bacteria size: diameter : 1micrometer, length: 5micrometer
'''
seed_num = 1
def Remove(duplicate): 
    final_list = [] 
    for num in duplicate: 
        if num not in final_list: 
            final_list.append(num) 
    return final_list
        
def x_binary_search(node, x_nodes):
    if len(x_nodes) == 1:
        return x_nodes[0]
    else:
        x = node[0]
        mid_1 = int(len(x_nodes)/2) - 1
        mid_2 = int(len(x_nodes)/2)
        dist_1 = ((x_nodes[mid_1] - x)**2)
        dist_2 = ((x_nodes[mid_2] - x)**2)
        if dist_1 < dist_2:
            return x_binary_search(node, x_nodes[:mid_2]) #since end is exclusive, while start is inclusive
        else:
            return x_binary_search(node, x_nodes[mid_2:])
        
def y_binary_search(node, y_nodes):
    if len(y_nodes) == 1:
        return y_nodes[0]
    else:
        y = node[1]
        mid_1 = int(len(y_nodes)/2) - 1
        mid_2 = int(len(y_nodes)/2)
        dist_1 = ((y_nodes[mid_1] - y)**2)
        dist_2 = ((y_nodes[mid_2] - y)**2)
        if dist_1 < dist_2:
            return y_binary_search(node, y_nodes[:mid_2]) #since end is exclusive, while start is inclusive
        else:
            return y_binary_search(node, y_nodes[mid_2:])

def Rotate(Particle, Surface, temp_dict, circle_list):
    y = 0 #key for lowest point
    min_val = 100
    for i in temp_dict:
            if i[2] < min_val:
                min_val = i[2]
                y = i
                
    bact_coord = temp_dict
    surf_coord = list(Surface.coordinates)
    temp_bact = copy.copy(temp_dict)
    #change how large of the bottom portion of the cylinder to measure
    search_range = 10  # NOTE! must be 1/4 of size of circle because it searches up and down
    energy_per_circle = []
    first_circle = 1
        
    #rotate labels of circle by 360 --> define as a function, reduces clutter and focuses output -->
    #needs to input: a particle and surface, 
    #needs to output: min_index, angle
    for circle in circle_list:
        num_rot = 0
        zlow = Particle.height
        zlow_key = 0
        for xyz in circle:
            if xyz[2] < zlow:
                zlow = xyz[2]
                zlow_key = xyz
        zlow_index = circle.index(zlow_key)
        energy_per_rot = []        
    #rotates around entire circle
        while num_rot <= len(circle): #rotates one unit around
            index = 0
            while index < len(circle):
                current_key = circle[index]
                if index == len(circle) - 1:
                    next_key = circle[0]
                else:
                    next_key = circle[index + 1]
                index += 1
                bact_coord[next_key] = temp_bact[current_key]
            temp_bact = copy.copy(bact_coord)
            num_rot += 1
    #calculate energy
            temp_energy = 0
            if zlow_index >= len(circle) - search_range:
                zlow_index = zlow_index - len(circle)
            for i in range(zlow_index - search_range, zlow_index + search_range + 1):
                temp_key = circle[i]
                x = temp_key[0]
                y = temp_key[1]
                z = temp_key[2]
                xy = (x, y, z)
                surf_key = (x_binary_search(xy, Surface.x), y_binary_search(xy, Surface.y), 0.0)
                r = temp_key[2] #height of each point which changes, causing energy to vary
                if Surface.coordinates[surf_key] == 'A' and bact_coord[temp_key] == 'A':
                    temp_energy += 1 / r #computes energy
                elif bact_coord[temp_key] == 'A' and Surface.coordinates[surf_key] == 'B' or Surface.coordinates[surf_key] == 'A' and bact_coord[temp_key] == 'B':
                    temp_energy += -1 / r
                elif bact_coord[temp_key] == 'B' and Surface.coordinates[surf_key] == 'B':
                    temp_energy += 1 / r
                elif bact_coord[temp_key] == 'C' and Surface.coordinates[surf_key] == 'C':
                    temp_energy += -(C*C) /r
                elif bact_coord[temp_key] == 'D' and Surface.coordinates[surf_key] == 'C' or Surface.coordinates[surf_key] == 'D' and bact_coord[temp_key] == 'C':
                    temp_energy += 0
                elif bact_coord[temp_key] == 'D' and Surface.coordinates[surf_key] == 'D':
                    temp_energy += -(D*D) /r
                else:
                    temp_energy += 1
            energy_per_rot.append(temp_energy)
        if first_circle == 1:
            energy_per_circle = energy_per_rot
            start_key = copy.copy(zlow_key)
            init_index = zlow_index
            first_circle += 1
        else:
            ind = 0
            while ind < len(energy_per_circle):
                energy_per_circle[ind] += energy_per_rot[ind] #ind is 27 error
                ind += 1
                
    #normalize: divide every energy_per_circle by 300, the number of interactions (15*20)
    norm = 0
    while norm < len(energy_per_circle):
        energy_per_circle[norm] = energy_per_circle[norm]/300
        norm += 1
        
    min_energy = min(energy_per_circle)
    min_index = energy_per_circle.index(min_energy)
    if (init_index + min_index) > len(Particle.circle_list[0]):
        rot_index = init_index + min_index - len(Particle.circle_list[0])
        angle = (abs(rot_index-init_index)/(Particle.num_coord-1))*360
    else:
        angle = (min_index/(Particle.num_coord-1))*360
    return(min_index, angle, min_energy)
                
        
x_val = []
for x in range(500):
    x_val.append(x*0.01)

random.seed(10)
np.random.seed(seed_num)
env = simpy.Environment()
character_list = ['A', 'B', 'C', 'D'] # (A - negative, B - positive, C - hydrophobic, D - hydrophilic)
#energy parameters
A = 1
B = -1
C = 1
D = 1

#Color-coordinated plot (before)
fig2 = plt.figure()    
ax2 = fig2.add_subplot(111, projection='3d')
# MUST write plt.show() after everything is drawn (not the best solution...)

#Color-coordinated plot (after interaction)
fig3 = plt.figure()    
ax3 = fig3.add_subplot(111, projection='3d')

class Particle(object):
    ''' A - negative, B - positive, C - hydrophobic, D - hydrophilic'''
    def __init__(self, radius, length, height, position, seed):
        self.r = radius
        self.coords = {}
        self.height = height
        self.length = length
        if position == 'R':
            self.position = random.choice(x_val)
        else:
            self.position = position
        self.num_coord = 73
        #makes cylinder      
        self.rad = [radius]
        num = [self.num_coord]
        circles = []
        for y in np.linspace(0, length, 20):
            for r , n in zip(self.rad, num):
                t = np.linspace(0, 2*math.pi, n)
                x = r * np.cos(t) + self.position
                z = r * np.sin(t) + self.height
                y = y * np.linspace(1, 1, n) + self.position
                circles.append(np.c_[np.round(x, 5), np.round(y, 5), np.round(z, 5)])
                
        self.circle_coords = []
        for i in circles:
            ct = 1
            for j in i:
                if ct % 73 != 0:
                    self.circle_coords.append(tuple(j))
                    ct += 1

        #forms rings of circles
        y = []
        self.circle_list = []
        #compile circle keys (x,y,z) sorted by y
        for xyz in self.circle_coords:
            if xyz[1] not in y:
                y.append(xyz[1])
                self.circle_list.append([xyz])
            else:
                index = y.index(xyz[1])
                self.circle_list[index].append(xyz)
        
        #print(len(self.circle_list), len(self.circle_coords))
        
        #AB dots
        
        for i in self.circle_coords:
            self.coords[i] = "A"
        
        #creating domains on cylinder (to change this, refer to GitHub readme)
        for i in range(36):
            index = np.random.randint(1, 1221)
            if (index - 71) % 72 == 0:
                index = index - 2
            elif (index - 70) % 72 == 0:
                index = index - 1
            elif (index % 72) == 0:
                index = index + 1
            #form octagons (circles)
            key_1 = self.circle_coords[index]
            key_2 = self.circle_coords[index+1]
            key_3 = self.circle_coords[index-1+72]
            key_4 = self.circle_coords[index+72]
            key_5 = self.circle_coords[index+1+72]
            key_6 = self.circle_coords[index+2+72]
            key_7 = self.circle_coords[index-1+144]
            key_8 = self.circle_coords[index+144]
            key_9 = self.circle_coords[index+1+144]
            key_10 = self.circle_coords[index+2+144]
            key_11 = self.circle_coords[index+216]
            key_12 = self.circle_coords[index+1+216]
            self.coords[key_1] = "B"
            self.coords[key_2] = "B"
            self.coords[key_3] = "B"
            self.coords[key_4] = "B"
            self.coords[key_5] = "B"
            self.coords[key_6] = "B"
            self.coords[key_7] = "B"
            self.coords[key_8] = "B"
            self.coords[key_9] = "B"
            self.coords[key_10] = "B"
            self.coords[key_11] = "B"
            self.coords[key_12] = "B"
            
        a = 0
        b = 0
        c = 0
        d = 0            
        for i in self.coords:
                if self.coords[i] == 'A':
                    a += 1
                elif self.coords[i] == 'B':
                    b += 1
                elif self.coords[i] == 'C':
                    c += 1
                else:
                    d += 1
        a_percent = a*100/(a+b+c+d)
        b_percent = b*100/(a+b+c+d)
        c_percent = c*100/(a+b+c+d)
        d_percent = d*100/(a+b+c+d)
        print("A: ", a_percent)
        print("B: ", b_percent)
        print("C: ", c_percent)
        print("D: ", d_percent)
        
        #3D Color coordinated (points-only)
        #Note: the patchiness is off angled to begin with, but doesn't matter if you rotate through all of it
        Ax = []
        Ay = []
        Az = []
        Bx = []
        By = []
        Bz = []
        Cx = []
        Cy = [] 
        Cz = []
        Dx = []
        Dy = []
        Dz = []

        for i in self.coords:
            if self.coords[i] == 'A':
                Ax.append(i[0])
                Ay.append(i[1])
                Az.append(i[2])
            elif self.coords[i] == 'B':
                Bx.append(i[0])
                By.append(i[1])
                Bz.append(i[2])
            elif self.coords[i] == 'C':
                Cx.append(i[0])
                Cy.append(i[1])
                Cz.append(i[2])
            else:
                Dx.append(i[0])
                Dy.append(i[1])
                Dz.append(i[2])
        
        ax2.scatter(Ax, Ay, Az, c = 'red', edgecolors='b', label = 'A')
        ax2.scatter(Bx, By, Bz, c = 'blue', edgecolors='b', label = 'B')
        ax2.scatter(Cx, Cy, Cz, c = 'green', edgecolors='b', label = 'C')
        ax2.scatter(Dx, Dy, Dz, c = 'black', edgecolors='b', label = 'D')
        ax2.legend(loc = "upper right")
        
        ax2.set_xlabel("X")
        ax2.set_ylabel("Y")
        ax2.set_zlabel("Z")
        #ax2.title.set_text("Number of Sims: " + str(self.num_sim))
        
        # Customize the z axis.
        ax2.set_zlim(0, 5)
        ax2.zaxis.set_major_locator(LinearLocator(10))
        ax2.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
        
        # rotate the axes and update
        for angle in range(0, 360):
           ax2.view_init(30, 40)
    
    def interact(self, Surface, search_range):
        #translation stuff
        #change the self.coords coordinates to different ones, and it should work. However, keep in mind the coordinate %system
        #self.r = 1, self.position = 2, bound is self.r to Surface.size - self.r
        x_bound = []
        y_bound = []
        increment_x = 21
        increment_y = 41 #how many intervals we move across the surface
        start_x = 0
        end_x = Surface.size_x 
        start = 3
        end = 13
        
        for i in np.linspace(start_x, end_x, increment_x):
            a = i - self.position
            x_bound.append(a)
        
        for i in np.linspace(start, end, increment_y):
            a = i - self.position
            y_bound.append(a)
        
        x_bound = [round(num,4) for num in x_bound]
        y_bound = [round(num,4) for num in y_bound]
        #print(x_bound, y_bound)
        
        translation_energy = []
        translation_index = []
        translation_angle = []
        
        #loop stuff
        #add loop through all x coords, and y coords addition to existing coordinates while marking all the
        counter = 0
        while counter < len(x_bound):
            counter2 = 0
            x = x_bound[counter]
            tempx = copy.copy(self.coords)
            newx = {}
            for coord in self.coords:
                old_key = copy.copy(coord)
                mod_key = copy.copy(coord)
                mod_key = list(mod_key)
                mod_key[0] += x
                mod_key = tuple(mod_key)
                newx[mod_key] = tempx[old_key]
            while counter2 < len(y_bound):
                y = y_bound[counter2]
                tempy = copy.copy(newx)
                newy = {}
                for coord in newx:
                    old_key = copy.copy(coord)
                    mod_key = copy.copy(coord)
                    mod_key = list(mod_key)
                    mod_key[1] += y
                    mod_key = tuple(mod_key)
                    newy[mod_key] = tempy[old_key]
                    
                tempy_labels = newy.keys()
                a = []
                circle_list = []
                #compile circle keys (x,y,z) sorted by y
                for xyz in tempy_labels:
                    if xyz[1] not in a:
                        a.append(xyz[1])
                        circle_list.append([xyz])
                    else:
                        index = a.index(xyz[1])
                        circle_list[index].append(xyz)

                circle_coords = []
                #form into one list
                for circle in circle_list:
                    for pos in circle:
                        circle_coords.append(pos)                    
                        
                #sorting into lines
                b = []
                line_list = []
                for coord in circle_coords:
                    if (coord[0], coord[2]) not in b:
                        b.append((coord[0], coord[2]))
                        line_list.append([coord])
                    else:
                        index = b.index((coord[0], coord[2]))
                        line_list[index].append(coord)
                
                rotation_energy = []
                rotation_index = []
                rotation_angle = []
                rotation_xy = []
                last_one = 1
                rotates = 17 #22.5 degrees per rot
                for y in np.linspace(0 , 2*math.pi, rotates):
                    if last_one != rotates:
                        last_one += 1
                        tempz = copy.copy(newy)
                        first_line = line_list[0]
                        x11 = first_line[0]
                        x12 = first_line[len(first_line)-1]
                        cx = (x11[0] + x12[0]) / 2
                        cy = (x11[1] + x12[1]) / 2
                        line1 = []
                        line_ct = 0
                        tempxyz = {}
                        for line in line_list:
                            #cx = (line[0][0] + line[len(line)-1][0]) / 2
                            #cy = (line[0][1] + line[len(line)-1][1]) / 2
                            for i in line:
                                x_new = ((i[0]-cx) * math.cos(y) - (i[1]-cy) * math.sin(y)) + cx
                                y_new = ((i[0]-cx) * math.sin(y) + (i[1]-cy) * math.cos(y)) + cy
                                new_key = (x_new, y_new, i[2])
                                tempxyz[new_key] = tempz[i]
                                if line_ct == 0:
                                    line1.append(new_key)
                            line_ct += 1
                        
                        tempz_labels = tempxyz.keys()
                        a = []
                        circle_list = []
                        #must sort by a diagonal line, hmmm
                        pt_1 = line1[0]
                        pt_last = line1[len(line1)-1]
                        if (round(pt_last[0],5) - round(pt_1[0],5)) == 0:
                            #compile circle keys (x,y,z) sorted by Y
                            for xyz in tempz_labels:
                                if round(xyz[1], 5) not in a:
                                    a.append(round(xyz[1], 5))
                                    circle_list.append([xyz])
                                else:
                                    index = a.index(round(xyz[1], 5))
                                    circle_list[index].append(xyz)
                        elif (round(pt_last[1],5) - (round(pt_1[1],5))) == 0:
                            #compile circle keys (x,y,z) sorted by X
                            for xyz in tempz_labels:
                                if round(xyz[0], 5) not in a:
                                    a.append(round(xyz[0], 5))
                                    circle_list.append([xyz])
                                else:
                                    index = a.index(round(xyz[0], 5))
                                    circle_list[index].append(xyz)
                        else:
                            perp_slope = (pt_last[1] - pt_1[1]) / (pt_last[0] - pt_1[0])
                            slope = - 1 / perp_slope
                            counted = 0
                            counted2 = 0
                            for pt in line1:
                                circle_list.append([pt])
                                counta = 0
                                for xyz in tempz_labels:
                                    if xyz != pt:
                                        temp_slope = round((xyz[1] - pt[1])/(xyz[0] - pt[0]), 6)
                                        #counted2 += 1
                                        #print(pt, list(tempz_labels)[1360], list(tempz_labels)[1419])
                                        #print(temp_slope, round(slope,6))
                                        #print(temp_slope, round(slope, 6)) #missing 40... 1420-1380
                                        if temp_slope == round(slope, 6) or math.isnan(temp_slope):
                                            circle_list[counted].append(xyz)
                                            counted2 += 1
                                    counta += 1
                                #print("woot", counta)
                                counted += 1
                            #print(len(circle_list), len(circle_list[0]))
                            #print("gucci", counted, counted2)
                        circle_coords = []
                        #form into one list
                        for circle in circle_list:
                            for pos in circle:
                                circle_coords.append(pos)
                                
                        #check if on surface (can make more efficient by checking 4 corners) 
                        #- gotta test if it continues that loop or bigger loop
                        #print(len(circle_coords), len(circle_list))
                        min_energy = 0
                        angle = 0
                        min_index = 0
                        if circle_coords[0][0] < 0 or circle_coords[0][0] > Surface.size_x or circle_coords[0][1] < 0 or circle_coords[0][1] > Surface.size_x:
                            rotation_energy.append(min_energy)
                            rotation_angle.append(angle)
                            rotation_index.append(min_index)
                            continue
                        if circle_coords[36][0] < 0 or circle_coords[36][0] > Surface.size_x or circle_coords[36][1] < 0 or circle_coords[36][1] > Surface.size_x:
                            rotation_energy.append(min_energy)
                            rotation_angle.append(angle)
                            rotation_index.append(min_index)
                            continue
                        if circle_coords[1368][0] < 0 or circle_coords[1368][0] > Surface.size_x or circle_coords[1368][1] < 0 or circle_coords[1368][1] > Surface.size_x:
                            rotation_energy.append(min_energy)
                            rotation_angle.append(angle)
                            rotation_index.append(min_index)
                            continue
                        if circle_coords[1404][0] < 0 or circle_coords[1404][0] > Surface.size_x or circle_coords[1404][1] < 0 or circle_coords[1404][1] > Surface.size_x:
                            rotation_energy.append(min_energy)
                            rotation_angle.append(angle)
                            rotation_index.append(min_index)
                            continue
                        #print(circle_coords[0][0], circle_coords[36][0], circle_coords[1368][0], circle_coords[1404][0],
                            # circle_coords[0][1], circle_coords[36][1], circle_coords[1368][1], circle_coords[1404][1])
                        
                        min_index, angle, min_energy = Rotate(self, Surface, tempxyz, circle_list)
                        rotation_energy.append(min_energy)
                        rotation_angle.append(angle)
                        rotation_index.append(min_index)

                translation_energy.append(rotation_energy) #can make this faster by keeping 
                #1 lowest energy and updating it as we go
                translation_index.append(rotation_index) #can make this faster by keeping a number instead of a list
                translation_angle.append(rotation_angle)
                counter2 += 1
            counter += 1
                
        
        #implement new way on how to find min translate and min rotate, and then changing the actual model
        ultimate_min = 0
        trans_index = 0
        rot_index = 0
        counter_trans = 0
        for rotation_set in translation_energy:
            counter_rot = 0
            for rotation in rotation_set:
                if rotation < ultimate_min:
                    ultimate_min = rotation
                    rot_index = counter_rot
                    trans_index = counter_trans
                    print(rot_index, trans_index)
                counter_rot += 1
            counter_trans += 1
        print("ultimate min: " , ultimate_min, "trans_energy: ", translation_energy)
        print(rot_index, trans_index)
        trans_rot_energy = translation_energy[trans_index][rot_index]
        print(translation_energy)
        trans_rot_angle = translation_angle[trans_index][rot_index]
        trans_rotx2_index = translation_index[trans_index][rot_index]

        x_index = trans_index // increment
        y_index = trans_index % increment
        add_x = x_bound[x_index]
        add_y = y_bound[y_index]
        
        trans_labels = self.coords.keys()
        new_labels = []
        for coord in trans_labels:
            coord = list(coord)
            coord[0] += add_x
            coord[1] += add_y
            coord = tuple(coord)
            new_labels.append(coord)
        
        values = []
        for key in self.coords:
            values.append(self.coords[key])
        
        trans_coords = {}
        ind = 0
        while ind < len(trans_labels):
            trans_coords[new_labels[ind]] = values[ind]
            ind += 1
                
        #rotating
        c = []
        line_list2 = []
        for coord in new_labels:
            if (coord[0], coord[2]) not in c:
                c.append((coord[0], coord[2]))
                line_list2.append([coord])
            else:
                index = c.index((coord[0], coord[2]))
                line_list2[index].append(coord)       
                
        rot_angles = np.linspace(0 , 2*math.pi, rotates)
        min_rot_angle = rot_angles[rot_index]
        newer_coords = copy.copy(trans_coords)
        first_line2 = line_list2[0]
        new_x11 = first_line2[0]
        new_x12 = first_line2[len(first_line2)-1]
        cx = (new_x11[0] + new_x12[0]) / 2
        cy = (new_x11[1] + new_x12[1]) / 2
        line2 = []
        line_ct = 0
        newest_coords = {}
        for line in line_list2:
            for i in line:
                x_new = ((i[0]-cx) * math.cos(min_rot_angle) - (i[1]-cy) * math.sin(min_rot_angle)) + cx
                y_new = ((i[0]-cx) * math.sin(min_rot_angle) + (i[1]-cy) * math.cos(min_rot_angle)) + cy
                new_key = (x_new, y_new, i[2])
                if line_ct == 0:
                    line2.append(new_key)
                newest_coords[new_key] = newer_coords[i]
            line_ct += 1
           
        d = []
        circle_list = []
        newer_labels = newest_coords.keys()
        #must sort by a diagonal line, hmmm
        pt_2 = line2[0]
        pt_last2 = line2[len(line2)-1]
        if round((pt_last2[0] - pt_2[0]), 5) == 0:
            #compile circle keys (x,y,z) sorted by Y
            for xyz in newer_labels:
                if round(xyz[1], 5) not in d:
                    d.append(round(xyz[1], 5))
                    circle_list.append([xyz])
                else:
                    index = d.index(round(xyz[1], 5))
                    circle_list[index].append(xyz)
        elif round((pt_last2[1] - pt_2[1]),5) == 0:
            print('no')
            #compile circle keys (x,y,z) sorted by X
            for xyz in newer_labels:
                if round(xyz[0], 5) not in d:
                    d.append(round(xyz[0], 5))
                    circle_list.append([xyz])
                else:
                    index = d.index(round(xyz[0], 5))
                    circle_list[index].append(xyz)
        else:
            print('yes')
            perp_slope2 = (pt_last2[1] - pt_2[1]) / (pt_last2[0] - pt_2[0])
            slope2 = - 1 / perp_slope2
            counter = 0
            for pt in line2:
                circle_list.append([pt])
                for xyz in newer_labels:
                    if xyz != pt:
                        temp_slope = round((xyz[1] - pt[1])/(xyz[0] - pt[0]), 6)
                        print(temp_slope, slope2)
                        if temp_slope == round(slope2, 6):
                            circle_list[counter].append(xyz)      
                counter += 1
        
        old_coord = newest_coords
        new_coord = copy.copy(newest_coords)
        for circle in circle_list:
            num_rot = 0        
            #rotates around entire circle
            while num_rot < trans_rotx2_index:
                #rotates one unit around
                index = 0
                while index < len(circle):
                    current_key = circle[index]
                    if index == len(circle) - 1:
                        next_key = circle[0]
                    else:
                        next_key = circle[index + 1]
                    index += 1
                    old_coord[next_key] = new_coord[current_key]
                new_coord = copy.copy(old_coord)
                num_rot += 1
        #print(new_coord)
        #counter for bottom portion of a,b,c,d
        a_s = 0
        b_s = 0
        c_s = 0
        d_s = 0

        for circle in circle_list:
            zlow = self.height
            zlow_key = 0
            for xyz in circle:
                if xyz[2] < zlow:
                    zlow = xyz[2]
                    zlow_key = xyz
            zlow_index = circle.index(zlow_key)
            for i in range(zlow_index - 3, zlow_index + 3 + 1):
                temp_key = circle[i]
                if new_coord[temp_key] == 'A':
                    a_s += 1
                elif new_coord[temp_key] == 'B':
                    b_s += 1
                elif new_coord[temp_key] == 'C':
                    c_s += 1
                elif new_coord[temp_key] == 'D':
                    d_s += 1
                    
        total = a_s + b_s + c_s + d_s
        print("A%: " + str(a_s/total), "B%: " + str(b_s/total), "C%: " + str(c_s/total), "D%: " + str(d_s/total))
        
        #calculate surface charge potential
        total_a = 0
        total_b = 0
        bact_a = 0
        bact_b = 0
        surf_key_list = []
        for circle in circle_list:
            zlow = self.height
            zlow_key = 0
            for xyz in circle:
                if xyz[2] < zlow:
                    zlow = xyz[2]
                    zlow_key = xyz
            zlow_index = circle.index(zlow_key)
            for i in range(zlow_index - 7, zlow_index + 7 + 1):
                temp_key = circle[i]
                surf_key = (x_binary_search(temp_key, Surface.x), y_binary_search(temp_key, Surface.y), 0.0)
                if surf_key not in surf_key_list:
                    surf_key_list.append(surf_key)
        
        for key in surf_key_list:
            if Surface.coordinates[key] == "A":
                total_a += 1
            elif Surface.coordinates[key] == "B":
                total_b += 1
            
        print("Total A: "+ str(total_a), "Total B: " + str(total_b))
             
        #3D Color coordinated (points-only)
        #Note: the patchiness is off angled to begin with, but doesn't matter if you rotate through all of it
        Ax = []
        Ay = []
        Az = []
        Bx = []
        By = []
        Bz = []
        Cx = []
        Cy = [] 
        Cz = []
        Dx = []
        Dy = []
        Dz = []
        Ex = []
        Ey = []
        Ez = []

        for i in new_coord:
            if new_coord[i] == 'A':
                Ax.append(i[0])
                Ay.append(i[1])
                Az.append(i[2])
            elif new_coord[i] == 'B':
                Bx.append(i[0])
                By.append(i[1])
                Bz.append(i[2])
            elif new_coord[i] == 'C':
                Cx.append(i[0])
                Cy.append(i[1])
                Cz.append(i[2])
            elif new_coord[i] == 'D':
                Dx.append(i[0])
                Dy.append(i[1])
                Dz.append(i[2])
            else:
                Ex.append(i[0])
                Ey.append(i[1])
                Ez.append(i[2])
        
        ax3.scatter(Ax, Ay, Az, c = 'red', edgecolors='b', label = 'A')
        ax3.scatter(Bx, By, Bz, c = 'blue', edgecolors='b', label = 'B')
        ax3.scatter(Cx, Cy, Cz, c = 'green', edgecolors='b', label = 'C')
        ax3.scatter(Dx, Dy, Dz, c = 'black', edgecolors='b', label = 'D')
        ax3.scatter(Ex, Ey, Ez, c = 'orange', edgecolors='b', label = 'E')
        ax3.legend(loc = "upper right")
                
        ax3.set_xlabel("X")
        ax3.set_ylabel("Y")
        ax3.set_zlabel("Z")
        ax3.title.set_text("SMALL AB Gradient seed " + str(105))
        #ax3.title.set_text("Num of sims: " + str(self.num_sim) + "  Angle: " + str(round(angle,3)) + " Min_energy:" + str(round(min_energy, 3)))
        #For differing plane surfaces
        #ax3.title.set_text("Num of sims: " + str(self.num_sim) + "  Angle: " + str(angle) + "  Plane:" + str(Surface.patchiness))
        
        # Customize the z axis.
        ax3.set_zlim(0, 5)
        ax3.zaxis.set_major_locator(LinearLocator(10))
        ax3.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
        
        # rotate the axes and update
        for angle in range(0, 360):
           ax3.view_init(30, 40)
                                   
class Surface(object):
    '''A - negative, B - positive, C - hydrophobic, D - hydrophilic'''
    def __init__(self, size_x, size_y):
        self.labels = []
        self.coordinates = {}
        self.size_x = size_x
        self.size_y = size_y
        points_x = 80
        points_y = 160
        np.random.seed(1)
        
        #makes surface
        self.x = np.linspace(0, self.size_x, points_x)
        self.y = np.linspace(0, self.size_y, points_y)
        
        xy, yx = np.meshgrid(self.x, self.y)
        xz = xy*0
        
        for i in range(len(xy)):
            for k in range(len(xy[i])):
                self.labels.append((xy[i][k], yx[i][k], xz[i][k]))
        
        #create a gradient from A --> B
        gradient = np.linspace(0, 1, points_y)
        line_ct = 0
        counter = 1
        firstlines = 0
        while counter <= len(self.labels):
            key = self.labels[counter-1]
            conc = [gradient[line_ct], 1 - gradient[line_ct]]
            if counter % points_x == 0:
                if firstlines < 50:
                    firstlines += 1
                elif line_ct < 99:
                    line_ct += 1
                #print(conc)
            if line_ct < 100:
                self.coordinates[key] = np.random.choice(['A', 'B'], size = 1, p = conc)
            else:
                self.coordinates[key] = "A"
            counter += 1
                     
        Ax = []
        Ay = []
        Az = []
        Bx = []
        By = []
        Bz = []
        Cx = []
        Cy = [] 
        Cz = []
        Dx = []
        Dy = []
        Dz = []
        
        for i in self.coordinates:
            if self.coordinates[i] == 'A':
                Ax.append(i[0])
                Ay.append(i[1])
                Az.append(i[2])
            elif self.coordinates[i] == 'B':
                Bx.append(i[0])
                By.append(i[1])
                Bz.append(i[2])
            elif self.coordinates[i] == 'C':
                Cx.append(i[0])
                Cy.append(i[1])
                Cz.append(i[2])
            else:
                Dx.append(i[0])
                Dy.append(i[1])
                Dz.append(i[2])
                
        ax2.scatter(Ax, Ay, Az, c = 'red')
        ax2.scatter(Bx, By, Bz, c = 'blue')
        ax2.scatter(Cx, Cy, Cz, c = 'green')
        ax2.scatter(Dx, Dy, Dz, c = 'black')
        ax3.scatter(Ax, Ay, Az, c = 'red')
        ax3.scatter(Bx, By, Bz, c = 'blue')
        ax3.scatter(Cx, Cy, Cz, c = 'green')
        ax3.scatter(Dx, Dy, Dz, c = 'black')
        
        ax2.set_xlabel("X")
        ax2.set_ylabel("Y")
        ax2.set_zlabel("Z")
        
        # Customize the z axis.
        ax2.set_zlim(0, 5)
        ax2.zaxis.set_major_locator(LinearLocator(10))
        ax2.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
        
        # rotate the axes and update
        for angle in range(0, 360):
           ax2.view_init(30, 40)

trial = Particle(0.5, 5, 0.501, 4, 1) #(radius, length, height, position, seed)
rando = Surface(10, 20) #changes the size of the surface (x,y)
trial.interact(rando, 7) #interaction function with the surface. (surface, search_range)
plt.show() #shows visualization
```
